```python

# Convert the date column to datetime objects
df['DateTime'] = pd.to_datetime(df['DateTime'], format="%d%b%Y")

# Extract only the date component
df['DateTime'] = df['DateTime'].dt.date

# Remove duplicate dates
df.drop_duplicates(subset=['DateTime'], inplace=True)
```

```python
df.head()
```

```python
# Convert the DateTime column to datetime objects
df['DateTime'] = pd.to_datetime(df['DateTime'])

# Resample the DataFrame to weekly intervals and keep the first record for each week
#df_weekly = df.resample('W-Mon', on='DateTime').first().reset_index()
df_weekly = df.resample('D', on='DateTime').sum().reset_index()

df_weekly = df_weekly.resample('M', on='DateTime').sum().reset_index()

# Print the DataFrame
print(df_weekly)
```

```python
# Assuming your DataFrame is called df and you want to export it to a file named 'output.csv'
df_weekly.to_csv('electricmontly.csv', index=False)
```
